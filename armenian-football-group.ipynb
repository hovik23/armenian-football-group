# Armenian Football Group Image Creating

Պահանջվող գրադարանների ներմուծում

```python
from PIL import Image, ImageDraw, ImageFont
```

## Ֆոնի ստեղծումը
Ֆոնի ստեղծումը պարունակում ա հետևյալ նկարի ընտրանքների նշումը՝

* Լայնություն
* Բարձրությունը
* Ֆոնի գույնը RGB ֆորմատով

Այս կերպ ստեղծվում ա ապագա նկարի հիմնքը։

```python
def create_background(width, height, color=((34,34,34))):
    img = Image.new('RGB', (width, height), color=color)
    return img

width, height = 1920, 950
background_color = (194,68,86)
img = create_background(width, height, color=background_color)
img
```

## Կողքի գծերի նկարչությունը

Հաջորդ քայլը սահմանային գծերի գծելն ա։ Այստեղ ֆունքցիայի ընտրանքների մեջ նշվում են՝

* Գծի լայնությունը
* Գծերի հեռավորությունը նկարի սահմաններից

```python
def add_lines(img, line_wdith=3, spacing=50):
    width, height = img.size
    draw = ImageDraw.Draw(img, 'RGBA')
    alpha = 200
    draw.line((spacing, spacing, width - spacing + line_wdith/2, spacing), fill=(255,255,255,alpha), width=line_wdith)
    draw.line((width - spacing, spacing + 1 + line_wdith/2, width - spacing, height - spacing), fill=(255,255,255,alpha), width=line_wdith)
    draw.line((width - spacing - line_wdith/2, height - spacing + 1 - line_wdith/2, spacing, height - spacing + 1 - line_wdith/2), fill=(255,255,255,alpha), width=line_wdith)
    draw.line((spacing + line_wdith/2, height - spacing - line_wdith, spacing + line_wdith/2, spacing + 1 + line_wdith/2), fill=(255,255,255,alpha), width=line_wdith)

    return img

img = add_lines(img, line_wdith=8)
img
```

## Լոգոի ավելացումը

Լոգոի ավելացումի նմանապես նշվուշ ա լոգոի անհրաժեշտ լայնությունը ու հեռավորությունը կողքի սահմաններից։ Բնականաբար, ընտանիքները կարված են նրանւց, թե որտեղ պիտի այդ նկարը կինի, ուստի այս դեպքում ընտրած են հետևյալ ընտրանքները։

```python
def add_logo(img, new_width=250, spacing_from_top=100):
    width, height = img.size
    logo = Image.open("./data/logo_main.png")
    wpercent = (new_width / logo.size[0])
    hsize = int((logo.size[1] * wpercent))
    logo = logo.resize((new_width, hsize))
#     img.paste(logo, (int((width - logo.size[0])/2), spacing_from_top), logo)
    img.paste(logo, (width - spacing_from_top - new_width, spacing_from_top), logo)
    return img

# img = add_logo(img, new_width=250, spacing_from_top=100)
img = add_logo(img, new_width=150, spacing_from_top=100)
img
```

## Ակումբների ավելցումը

Այստեղ կատարվում ա երևի ամենակարևոր մասը։ Նախորոք ստեղծվուկ ա տվյալների բազա, որը պարունակում ա ակումբների անունը երեք լեզվով (Հայերեն, Անգլերեն և Ռուսերեն) ու նշանը (լոգոն)։ Իմիջայլոց, քանի որ այս ծրագրի հետ աշխատանքը իրականացվելու ա Telegram բոտի միջոցով, այնտեղ իրականացված ա տվյալների բազայի հետ աշխատելու համայարք։ Մի խոսքով, հնարավոր ա ավելացնել  և ջնջել ակումբներ։ Սա տալիս ա լավ հնարավորություն տվայլների բազայի հետ աշխատելու համաը։
Վերադառնալով, ծրագրի այս մասին, ահստեղ ընտրվում են երկու ակումբ տվյալների բազայից, որոնց չափերը հավասարվում են։ Ավելի ճիշտ, հավասարվում են նշանների բարձրությունները, որովհետև ոնց ցույց տվեց պրակտիկան, վերջնական նկարը ավելի սիրոյն ա ստացվում, երբ որ ակումբների նշանները ունեն նույն բարձրությունը, ոչ թե լայնությունը։

```python
def add_teams(img, home_team, away_team, team_font=100):
    width, height = img.size
    home_team = home_team.upper()
    away_team = away_team.upper()
    
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype('./data/fonts/arnamu_bold.ttf', team_font)
    text = home_team + ' - ' + away_team
    
    home_team_width, home_team_height = draw.textsize(home_team, font=font)
    away_team_width, away_team_height = draw.textsize(away_team, font=font)
    
    spacing = 200
    spacing_name = 30
    team_y = spacing + 450 + spacing_name
    
    dash_width, dash_height = draw.textsize('-', font=font)
    team_width = (width - dash_width)/2
    
    draw.text( ((team_width - home_team_width)/2, team_y), home_team, font=font, fill="white")
    draw.text( ((team_width - away_team_width)/2 + team_width + dash_width, team_y), away_team, font=font, fill="white")
    
    home_team_logo = Image.open("./data/img/1.png")
    hpercent = home_team_logo.size[1] / 450
    wsize = int(home_team_logo.size[0] / hpercent)
    home_team_logo = home_team_logo.resize((wsize, 450))
    
    away_team_logo = Image.open("./data/img/4.png")
    hpercent = away_team_logo.size[1] / 450
    wsize = int(away_team_logo.size[0] / hpercent)
    away_team_logo = away_team_logo.resize((wsize, 450))
    
    team_logo_y = 200
    
    
    dash_y = team_logo_y + home_team_logo.size[1]/2 - dash_height/2
    draw.text(( (width - dash_width)/2, dash_y), '-', font=font, fill="white")
    
    
    img.paste(home_team_logo, (int( (team_width - home_team_logo.size[0])/2 ), team_logo_y), home_team_logo)
    img.paste(away_team_logo, (int( (team_width - away_team_logo.size[0])/2 + team_width + dash_width ), team_logo_y), away_team_logo)

    return img

img = add_teams(img, 'Alashkert', 'BKMA')
img
```

## Արդյունքի ավելացումը

Հետևյալ մասերը իրականացված են նույն ձև

```python
def add_result(img, home_goals, away_goals):
    width, height = img.size
    draw = ImageDraw.Draw(img, 'RGBA')
    font = ImageFont.truetype('./data/fonts/arnamu.ttf', 250)
    num_width, num_height = draw.textsize('1', font=font)
    dash_y = 400 - num_height/2
    spacing = 50
    draw.text( (width/2 - spacing - num_width, dash_y), str(home_goals), font=font, fill="white")
    draw.text( (width/2 + spacing, dash_y), str(away_goals), font=font, fill="white")
    return img

img = add_result(img, 1, 2)
img
```

## Ամսաթվի ավելացումը

```python
def add_time(img, time):
    width, height = img.size
    draw = ImageDraw.Draw(img, 'RGBA')
    font = ImageFont.truetype('./data/fonts/arnamu_bold.ttf', 60)
    time_width, time_height = draw.textsize(time, font=font)
    
    draw.text( ((width - time_width)/2, 200 + 380), time, font=font, fill="white")
    return img

img = add_time(img, '23/04/2022')
img
```

## Մրցակցության ավելացումը

```python
def add_tournament(img, new_width=250, spacing=100):
    width, height = img.size
    logo = Image.open("./data/icons/vbet_bw.png")
    wpercent = (new_width / logo.size[0])
    hsize = int((logo.size[1] * wpercent))
    logo = logo.resize((new_width, hsize))
    img.paste(logo, (spacing, spacing), logo)
    return img

img = add_tournament(img, new_width=150, spacing=100)
img
```

## Լրացուցիչ ինֆորմացիայի ավելացումը

Ժամանակավոր լուծում՝ եթե պայմանը կատարվում ա, ապա նկարի բարձրությունը ընտրվում ա որպես ֆիքսված թիվ, հակառակ դեպքում նկարի բարձրությունը հաշվում ա, ֆուտբոլիստների քանակից կախված

```python
def add_scorer(img, home_scorers, away_scorers):
    width, height = img.size
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype('./data/fonts/arnamu.ttf', 40)
    test_width, test_height = draw.textsize('Սահակյան'.upper(), font=font)
    
    spacing = 200
    ball_spacing = 10
    scorers_spacing = 10
    for i in range(len(home_scorers)):
        ball_icon = Image.open("./data/icons/ball_icon.png").convert("RGBA")
        wpercent = test_height / ball_icon.size[0]
        hsize = int((ball_icon.size[1] * wpercent))
        ball_icon = ball_icon.resize((test_height, hsize))
        
        home_scorers[i] = home_scorers[i].upper()
        scorer_width, scorer_height = draw.textsize(home_scorers[i], font=font)
        img.paste(ball_icon, (spacing, 850 + i * (scorers_spacing + test_height)), ball_icon)
        draw.text( (spacing + ball_spacing + ball_icon.size[0], 850 + i * (scorers_spacing + test_height)), str(home_scorers[i]), font=font, fill="white")
        
    for i in range(len(away_scorers)):
        ball_icon = Image.open("./data/icons/ball_icon.png").convert("RGBA")
        wpercent = test_height / ball_icon.size[0]
        hsize = int((ball_icon.size[1] * wpercent))
        ball_icon = ball_icon.resize((test_height, hsize))
        
        away_scorers[i] = away_scorers[i].upper()
        scorer_width, scorer_height = draw.textsize(away_scorers[i], font=font)
        img.paste(ball_icon, (width - spacing - ball_icon.size[0], 850 + i * (scorers_spacing + test_height)), ball_icon)
        draw.text( (width - spacing - ball_spacing - ball_icon.size[0] - scorer_width, 850 + i * (scorers_spacing + test_height)), str(away_scorers[i]), font=font, fill="white")

    return img

home_scorers = ["Սահակյան 23'"]
away_scorers = ["Սահակյան 24'", "Սահակյան 25'"]
img_width = 1920

if 1 == 9:
    img_height = 200 + 450 + 30 + 200 + 80
else:
    scorers_space = 10
    text_height = 39
    img_height = 200 + 450 + 30 + 200 + (text_height + scorers_space) * max(len(home_scorers), len(away_scorers)) + 80
    
img = create_background(img_width, img_height, color=(196,68,94))
img = add_teams(img, 'Alashkert', 'BKMA')
img = add_logo(img, new_width=150, spacing_from_top=100)
img = add_lines(img, line_wdith=8)
img = add_result(img, 2, 2)
img = add_tournament(img, new_width=150, spacing=100)
img = add_scorer(img, home_scorers, away_scorers)
# img.save('result.png')
img
```

# Լրացուցիչ նկարներ

```python
def add_goal_team(img, team, text):
    width, height = img.size
    
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype('./data/fonts/arnamu_bold.ttf', 150)
    text_width, text_height = draw.textsize(text.upper(), font=font)

    team_logo = Image.open("./data/img/" + str(team) +".png")
    hpercent = team_logo.size[1] / 450
    wsize = int(team_logo.size[0] / hpercent)
    team_logo = team_logo.resize((wsize, 450))

    spacing = 60
    img.paste(team_logo, ((int((width/2 - (team_logo.size[0] + spacing + text_width)/2))), int(height/2 - team_logo.size[1]/2)), team_logo)

    width_1 = width/2 - (team_logo.size[0] + spacing + text_width)/2 + team_logo.size[0] + spacing
    height_1 = height/2 - text_height/2
    draw.text((width_1, height_1), text.upper(), font=font, fill="white")
    
    return img

img = create_background(1920, 800, color=(23,23,23))
img = add_lines(img, line_wdith=8)
img = add_goal_team(img, 3, 'ԳՈՈՈՈՈՈՈԼ!')
img = add_logo(img, new_width=150, spacing_from_top=100)
img
```
